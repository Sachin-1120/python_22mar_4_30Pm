```python
def fine(a):
    if student[a]['Issued_book']!=None:
        submission=datetime.date.today()
        issue=student[a]['Issued_book']['Date_of_issue']
        days=(submission-issue).days
        if days>7:
            return student[a]['Fine']+(days-7)*2
        else:
            return student[a]['Fine']
    else:
        return student[a]['Fine']

def login():
    while True:
        n=int(input('''
        Press 1 for Login
        Press 2 For Exit
    
        '''))
        if n==1:
            a=input('Enter Your Student_id')
            if a in student.keys():
                n=int(input('''
                Press 1 for Issue Book
                Press 2 to submit Book
                Press 3 to submit Fine
                Press 4 to check Fine
                Press 5 for check Personal Info
                Press 6 to Exit
                '''))
                if n==1:
                    book_id=input('Enter Book Id to Issue')
                    if book_id in book.keys():
                        if book[book_id]['copies']>=1:
                            print('Book Issued successfully')
                            print('Kindly Resubmit Book in 7 days otherwise fine will be imposed')
                            book[book_id]['copies']-=1

                            student[a]['Issued_book']['Book_id']=book_id
                            student[a]['Issued_book']['name']=book[book_id]['title']
                            student[a]['Issued_book']['Date_of_issue']=datetime.date.today
                            student[a]['Fine']=0
                elif n==2:
                    book_id=a['Issued_book']['Book_id']
                    print(f'{book_id} Submitted Successfully')
                    Fine=fine(a)
                    a['Fine']+=Fine
                    a['Issued_Book']=None
                    book[book_id]['copies']+=1
                elif n==3:
                    print(f'Total Fine Till Today -->{fine(a)}')
                    n=int(input('Enter Amount To Submit'))
                    student[a]['Fine']-=n
                    
                elif n==4:
                    print(f'Fine of studentId-{a} is {fine(a)}')
                elif n==5:
                    print(f'''
                    Hello {student[a]["name"]}
                    
                        Your Contact Detail is :{student[a]['contact']}
                        Your student Id is     : {a}
                    ''')
                elif n==6:
                    sys.exit('thanks For the Service')   
                
            else:
                print('No Such Student Found')
        elif n==2:
            sys.exit('User Requested to exit')
        else:
            print('Invalid Input')
```
